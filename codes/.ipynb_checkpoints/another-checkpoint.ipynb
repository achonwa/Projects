```python
def hello(name= "DAN"):
    print("The  hello() function has been executed!")
    
    def greet():
        return '\t This is the greet() func inside hello!'
    
    def welcome():
        return '\t This is welcome() inside hello'
    
    if name == 'DAN':
        return greet
    else:
        return welcome
```

```python
mynfunc=hello("DAN")
```

```python
print(mynfunc())
```

```python
def new_decoratrs(original_func):
    
    def wrap_func():
        
        print("Some extra code ")
        
        original_func()
        
        print("Some extra code ")
        
    return wrap_func
        
```

```python
@new_decoratrs
def fundecorate():
    print("This has been decorated")
```

```python
fundecorate()
```

```python
## TIMING YOUR CODE

import time
```

```python
def func_one(n):
    return [str(num) for num in range(n)]
```

```python
func_one(14)
```

```python
def func_two(n):
    return list(map(str,range(n)))
```

```python
func_two(14)
```

```python
start_time = time.time()

result = func_one(100)

end_time = time.time()

elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
start_time = time.time()

result = func_two(100)

end_time = time.time()

elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
import timeit
```

```python
timeit.timeit
```

```python
stmt = '''
func_one(100)
'''
```

```python
setup = '''
def func_one(n):
    return [str(num) for num in range(n)]
'''
```

```python
timeit.timeit(stmt,setup,number=1000000)
```

```python
stmt2 = '''
func_two(100)
'''
```

```python
setup2 = '''
def func_two(n):
    return list(map(str,range(n)))
'''
```

```python
timeit.timeit(stmt2,setup2,number=1000000)
```

```python
%%timeit
func_one(100)
```

```python
%%timeit
func_two(100)
```
